```python
sec_list = [1,2,3]
sec =random.choice(sec_list)
ua = UserAgent()
headers = ua.random
fail_list =[]
for index,row in tqdm(style_img.iterrows()):
    # no = img_no[0] 從圖片網址提取編號
    # img_no = re.findall('.*products/(.*)/.*',img_url)
    no = row[0]
    style_name = row[1]
    tag_code = row[2]
    img_url = row[5]   
    img_jpg = requests.get(img_url,headers=headers)
    if tag_code == "E1_1":
        with open("素材風格(主圖)/率性休閒 E1_1/" + no + ".jpg", "ab") as file:  # 開啟資料夾及命名圖片檔 wb:假如存在就刪除覆蓋 ab:接著後面append
            file.write(img_jpg.content)
    elif tag_code == "E1_2":
        with open("素材風格(主圖)/簡約質感 E1_2/" + no + ".jpg", "ab") as file:  
            file.write(img_jpg.content)
    elif tag_code == "E1_3":
        with open("素材風格(主圖)/清新甜美 E1_3/" + no + ".jpg", "ab") as file:
            file.write(img_jpg.content)
    elif tag_code == "E1_4":
        with open("素材風格(主圖)/都會輕熟 E1_4/" + no + ".jpg", "ab") as file:
            file.write(img_jpg.content)
    elif tag_code == "E1_5":
        with open("素材風格(主圖)/溫柔浪漫 E1_5/" + no + ".jpg", "ab") as file:
            file.write(img_jpg.content)
    elif tag_code == "E1_6":
        with open("素材風格(主圖)/慵懶度假 E1_6/" + no + ".jpg", "ab") as file:
            file.write(img_jpg.content)
    elif tag_code == "E1_7":
        with open("素材風格(主圖)/氣質優雅 E1_7/" + no + ".jpg", "ab") as file:
            file.write(img_jpg.content)
    elif tag_code == "E1_8":
        with open("素材風格(主圖)/性感迷人 E1_8/" + no + ".jpg", "ab") as file:
            file.write(img_jpg.content)
    elif tag_code == "E1_9":
        with open("素材風格(主圖)/態度個性 E1_9/" + no + ".jpg", "ab") as file:
            file.write(img_jpg.content)
    elif tag_code == "E1_10":
        with open("素材風格(主圖)/其他 E1_10/" + no + ".jpg", "ab") as file:  
            file.write(img_jpg.content)
    elif tag_code == "E1_11":
        with open("素材風格(主圖)/時髦性感 E1_11/" + no + ".jpg", "ab") as file:
            file.write(img_jpg.content)
    else:
        fail_list.append(index)
        print(f"第{index}張圖片沒有下載成功")
    
    time.sleep(sec)

    
```
